```python
ts = df[['State Code', 'County Code', 'Site Num', 'Date Local', 'O3 Mean', 'NO2 Mean', 'SO2 Mean', 'CO Mean']].groupby(['Site Num', 'Date Local', 'State Code', 'County Code', ]).agg({
    'O3 Mean': 'mean',
    'NO2 Mean': 'mean',
    'SO2 Mean': 'mean', 
    'CO Mean': 'mean'
}).reset_index()
```

```python
# ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
site = 1
# types = ['O3 Mean', 'NO2 Mean', 'SO2 Mean', 'CO Mean']
types = ['CO Mean']
temp  = ts[ts['Site Num']==site][types]
start_date = ts[ts['Site Num']==site].iloc[0]['Date Local']

temp = temp.set_index(pd.date_range(start_date, periods=temp.shape[0]))
temp.plot(figsize=(18,5));
```
